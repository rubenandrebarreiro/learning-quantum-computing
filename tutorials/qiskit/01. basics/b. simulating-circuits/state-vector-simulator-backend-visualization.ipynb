```python
# Import NumPy as 'NP'
import numpy as np

# Import all libraries and tools from QisKit
from qiskit import *

# Using MatPlotLibrary for Python
%matplotlib inline
```

```python
# Create a Quantum Register with 3 qubits,
# with the name of 'my_quantum_register'.
quantum_register = QuantumRegister(3, 'my_quantum_register')

# Create a Quantum Circuit acting on the 'my_quantum_register' register
quantum_circuit = QuantumCircuit(quantum_register)
```

```python
# Add a H gate on qubit 0,
# putting this qubit in superposition
quantum_circuit.h(quantum_register[0])

# Add a CX (C_NOT) gate on control qubit 0 and target qubit 1,
# putting the qubits in a Bell state
quantum_circuit.cx(quantum_register[0], quantum_register[1])

# Add a CX (C_NOT) gate on control qubit 0 and target qubit 2,
# putting the qubits in a GHZ state

# Note:
# - This is a C_NOT Gate applied in a Circuit's Address,
#   so, may vary between executions
quantum_circuit.cx(quantum_register[0], quantum_register[2])
```

```python
# Run the Quantum Circuit on a State Vector Simulator's Backend
backend = BasicAer.get_backend('statevector_simulator')
```

```python
# Create a Quantum Program for execution of the Quantum Circuit
job = execute(quantum_circuit, backend)
```

```python
# When you run a program,
# a job object is made that
# has the following two useful methods:

# 1) job.status()
# 2) job.result()

# which return the status of
# the job and a result object respectively
result = job.result()
```

```python
# The results object contains the data
# and QisKit provides the method result.get_statevector(quantum_circuit)
# to return the state vector for the Quantum Circuit
output_state = result.get_statevector(quantum_circuit, decimals = 3)
print(output_state)
```

```python
# QisKit also provides a visualization toolbox to
# allow you to view these results

# Below, we use the visualization function to
# plot the real and imaginary components of the state vector
from qiskit.tools.visualization import plot_state_city

# Draw the plot for the
# real and imaginary components of the state vector
plot_state_city(output_state)
```
